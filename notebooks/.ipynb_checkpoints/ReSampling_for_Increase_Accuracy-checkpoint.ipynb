```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import LabelEncoder

# Load dataset
df_old = pd.read_csv("../data/raw/crop_and_soil_dataSet_resampled_new.csv")

# Separate features and target (assuming 'Crop Type' is the target)
X = df_old.drop(columns=["Crop Type", "Soil Type"])  # Exclude 'Crop Type' and 'Soil Type' columns from features
y = df_old["Crop Type"]  # Set 'Crop Type' as the target column

# Identify and encode categorical columns in features
categorical_columns = X.select_dtypes(include=['object']).columns

# Initialize the label encoder
label_encoder = LabelEncoder()

# Apply label encoding to categorical columns in X
encoding_map_X = {}  # Store label encoders for inverse transformation later
for col in categorical_columns:
    X[col] = label_encoder.fit_transform(X[col])
    encoding_map_X[col] = label_encoder  # Store the label encoder for this column

# If target column is categorical, encode it as well
y_encoded = label_encoder.fit_transform(y)

# Apply SMOTE to balance the dataset
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y_encoded)

# Reverse encoding for the features (X)
for col in categorical_columns:
    X_resampled[col] = encoding_map_X[col].inverse_transform(X_resampled[col].astype(int))

# Convert resampled target variable back to original labels
y_resampled_original = label_encoder.inverse_transform(y_resampled)

# Add "Soil Type" back as a string (assuming it's categorical)
df_resampled = pd.DataFrame(X_resampled, columns=X.columns)
df_resampled["Crop Type"] = y_resampled_original  # Add original target back to DataFrame

# Add "Soil Type" as a string (it can be encoded later if needed)
df_resampled["Soil Type"] = df_old["Soil Type"].iloc[0]  # Assuming a single value for 'Soil Type' in all rows

# Save the resampled dataset
df_resampled.to_csv("../data/processed/crop_and_soil_dataSet_resampled.csv", index=False)

print("Resampling completed! The new dataset is saved at ../data/processed/crop_and_soil_dataSet_resampled.csv")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load dataset
df_old = pd.read_csv("../data/raw/crop_and_soil_dataSet_resampled_new.csv")

# Define features and target
x1_feature = ['Temparature', 'Humidity', 'Moisture', 'Nitrogen', 'Potassium', 'Phosphorous', 'Soil Type']
y1_feature = ['Crop Type']

# Print the features and target
print(x1_feature)
print(y1_feature)

# Extract features and target columns
x1 = df_old.loc[:, x1_feature].values
y1 = df_old.loc[:, y1_feature].values

print("X1 Shape: ", x1.shape)
print("Y1 Shape: ", y1.shape)

# Initialize LabelEncoder for categorical features
label_encoder = LabelEncoder()

# Apply label encoding to 'Soil Type' (the categorical feature)
df_old['Soil Type'] = label_encoder.fit_transform(df_old['Soil Type'])

# Now, update the x1 with encoded 'Soil Type'
x1 = df_old.loc[:, x1_feature].values

print("Updated X1 Shape: ", x1.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x1, y1, test_size=0.2)


print("X_train shape : " , X_train.shape)
print("X_test shape : " , X_test.shape)
print("Y_train shape : " , Y_train.shape)
print("Y_test shape : " , Y_test.shape)
```

```python
from sklearn.metrics import accuracy_score  # Make sure this import is included
from sklearn.tree import DecisionTreeClassifier, export_text

# Use Decision Tree
decisionTree_model = DecisionTreeClassifier(random_state=0, max_depth=10, min_samples_split=10, min_samples_leaf=5)
decisionTree_model = decisionTree_model.fit(X_train, Y_train)

# Export the decision tree as text
# tree = export_text(decisionTree_model, feature_names=x1_feature)
# print(tree)

# Predict and compute accuracy
Y_pred = decisionTree_model.predict(X_test)
# print(Y_pred)

print("Train Accuracy Model_DT: ", decisionTree_model.score(X_train, Y_train))
dt_accuracy = accuracy_score(Y_test, Y_pred)  # Use accuracy_score here

print("Test Accuracy: ", dt_accuracy)
print(f'Accuracy of the Decision tree model on the test set: {dt_accuracy * 100:.2f}%')
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from xgboost import XGBClassifier
import pandas as pd

# Load new dataset
df = df_old.copy()

# Features and target
X = df.drop('Crop Type', axis=1)
Y = df['Crop Type']

# Encode categorical variables
le_soil = LabelEncoder()
le_fert = LabelEncoder()
X['Soil Type'] = le_soil.fit_transform(X['Soil Type'])
X['Fertilizer Name'] = le_fert.fit_transform(X['Fertilizer Name'])

# Encode target
le_y = LabelEncoder()
Y_encoded = le_y.fit_transform(Y)

# Scale numerical features
scaler = StandardScaler()
X[['Temparature', 'Humidity', 'Moisture', 'Nitrogen', 'Potassium', 'Phosphorous']] = scaler.fit_transform(
    X[['Temparature', 'Humidity', 'Moisture', 'Nitrogen', 'Potassium', 'Phosphorous']]
)

# Split data (16000 train, 4000 test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_encoded, test_size=0.2, random_state=0)

# Train XGBoost with adjusted settings
xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,            # Shallower to reduce overfitting
    learning_rate=0.05,     # Slower learning for generalization
    random_state=0,
    n_jobs=-1,
    objective='multi:softmax',
    num_class=11
)
xgb_model.fit(X_train, Y_train)

# Predict and compute accuracy
Y_pred = xgb_model.predict(X_test)
train_accuracy = xgb_model.score(X_train, Y_train)
test_accuracy = accuracy_score(Y_test, Y_pred)

print("Train Accuracy Model_XGB: ", train_accuracy)
print("Test Accuracy: ", test_accuracy)
print(f'Accuracy of the XGBoost model on the test set: {test_accuracy * 100:.2f}%')

# Cross-validation
cv_scores = cross_val_score(xgb_model, X, Y_encoded, cv=5)
print("5-Fold Cross-Validation Accuracy: ", cv_scores.mean(), "+/-", cv_scores.std())

# Feature importance
importances = xgb_model.feature_importances_
feature_names = X.columns
for name, importance in zip(feature_names, importances):
    print(f"{name}: {importance:.4f}")
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from xgboost import XGBClassifier
import pandas as pd

# Load new dataset
df = pd.read_csv('crop_dataset_20000_v2.csv')

# Features and target
X = df.drop('Crop Type', axis=1)
Y = df['Crop Type']

# Encode categorical variables
le_soil = LabelEncoder()
le_fert = LabelEncoder()
X['Soil Type'] = le_soil.fit_transform(X['Soil Type'])
X['Fertilizer Name'] = le_fert.fit_transform(X['Fertilizer Name'])

# Encode target
le_y = LabelEncoder()
Y_encoded = le_y.fit_transform(Y)

# Scale numerical features
scaler = StandardScaler()
X[['Temparature', 'Humidity', 'Moisture', 'Nitrogen', 'Potassium', 'Phosphorous']] = scaler.fit_transform(
    X[['Temparature', 'Humidity', 'Moisture', 'Nitrogen', 'Potassium', 'Phosphorous']]
)

# Split data (16000 train, 4000 test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_encoded, test_size=0.2, random_state=0)

# Train XGBoost with adjusted settings
xgb_model = XGBClassifier(
    n_estimators=300,
    max_depth=6,            # Shallower to reduce overfitting
    learning_rate=0.05,     # Slower learning for generalization
    random_state=0,
    n_jobs=-1,
    objective='multi:softmax',
    num_class=11
)
xgb_model.fit(X_train, Y_train)

# Predict and compute accuracy
Y_pred = xgb_model.predict(X_test)
train_accuracy = xgb_model.score(X_train, Y_train)
test_accuracy = accuracy_score(Y_test, Y_pred)

print("Train Accuracy Model_XGB: ", train_accuracy)
print("Test Accuracy: ", test_accuracy)
print(f'Accuracy of the XGBoost model on the test set: {test_accuracy * 100:.2f}%')

# Cross-validation
cv_scores = cross_val_score(xgb_model, X, Y_encoded, cv=5)
print("5-Fold Cross-Validation Accuracy: ", cv_scores.mean(), "+/-", cv_scores.std())

# Feature importance
importances = xgb_model.feature_importances_
feature_names = X.columns
for name, importance in zip(feature_names, importances):
    print(f"{name}: {importance:.4f}")
```
