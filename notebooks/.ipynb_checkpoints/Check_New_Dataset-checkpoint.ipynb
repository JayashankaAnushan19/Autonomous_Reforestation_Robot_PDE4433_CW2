```python
import pandas as pd

# Path to the Excel file
# https://data.mendeley.com/datasets/8v757rr4st/1
file_name = "../data/raw/Crop Recommendation using Soil Properties and Weather Prediction (1)_.csv"

# Load all sheets into a dictionary of DataFrames
df = pd.read_csv(file_name)

df.head()
```

```python
df.shape
```

```python
# df[df["Soilcolor"] == 'red'].value_counts()
# df[df["Soilcolor"].isin(["red", "Red", "Sandy", "sandy", "Loamy", "loamy", "Clayey", "clayey", "Black", "black"])].shape
new_df=df[df["Soilcolor"].isin(["red", "Red", "Sandy", "sandy", "Loamy", "loamy", "Clayey", "clayey", "Black", "black"])]
new_df.shape
new_df["Soilcolor"].value_counts()
```

```python
new_df.columns
```

```python
new_df.isna().sum()
```

```python
new_df.dtypes
```

```python
new_df.loc[:, 'Temperature'] = new_df[['T2M_MAX-W', 'T2M_MAX-Sp', 'T2M_MAX-Su', 'T2M_MAX-Au', 
                                       'T2M_MIN-W', 'T2M_MIN-Sp', 'T2M_MIN-Su', 'T2M_MIN-Au']].mean(axis=1)

new_df.loc[:, 'Humidity'] = new_df[['QV2M-W', 'QV2M-Sp', 'QV2M-Su', 'QV2M-Au']].mean(axis=1)
```

```python
new_df.head()
```

```python
# new_df[new_df["Soilcolor"]=="Red"].hist(bins=100, figsize=(50,50))
new_df.hist(bins=100, figsize=(50,50))
```

```python
import matplotlib.pyplot as plt

# Select 50% of the samples randomly
sampled_df = new_df.sample(frac=0.5, random_state=42)  # Adjust random_state for reproducibility

# Plot histogram
sampled_df.hist(bins=100, figsize=(20, 20))  # Adjust figsize if needed
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select first 50% of the data
half_df = new_df.iloc[:len(new_df) // 4]  # First 50% of rows

# Plot histogram
half_df.hist(bins=100, figsize=(20, 20))  # Adjust figsize if needed
plt.show()
```
